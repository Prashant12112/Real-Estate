```python
def treat_price(price_str):
    if pd.isna(price_str):
        return np.nan
    price_str = str(price_str).replace(',', '') # Remove commas
    parts = price_str.split(' ')

    if len(parts) == 2:
        try:
            value = float(parts[0])
            unit = parts[1]
            if unit == 'Lac':
                return round(value / 100, 2) # Convert Lac to Crore
            elif unit == 'Crore':
                return round(value, 2)
            else:
                return np.nan # Unknown unit
        except ValueError:
            return np.nan # Could not convert value to float
    else:
        # Attempt to convert directly if no unit or different format
        try:
            return round(float(price_str), 2)
        except ValueError:
            return np.nan # Cannot parse
```

```python
df['price'] = df['price'].apply(treat_price)
df.head()
```

```python
df['price_per_sqft'] = df['price_per_sqft'].apply(lambda x: float(str(x).replace('₹', '').replace('/sq.ft.', '').replace(',', '').strip()) if pd.notna(x) else np.nan)
```
